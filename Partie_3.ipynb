```python
import rdflib

g = rdflib.Graph()
g.parse("movies_project.owl")
iri = "https://schema.org/"
```

List the instances of the class Person

```python
sparql = """SELECT ?a 
        WHERE {
                ?a a <https://schema.org/myPerson>  .
                
                }"""

res = g.query(sparql)
sum(1 for _ in res)
for row in res:
    print(row.a)
```

List the name of all Thriller movies. For each one, display its director.

```python
sparql = """SELECT ?m_title ?d_name
        WHERE {
            ?m <https://schema.org/hasGenre> <https://schema.org/Thriller> .
            ?d <https://schema.org/isDirector> ?m .
            ?m <https://schema.org/mytitle> ?m_title .
            ?d <https://schema.org/myname> ?d_name
        }"""

res = g.query(sparql)
for row in res:
    print(f"Movie : {row.m_title} \t Director : {row.d_name}")
```

List the name of all Crime Thriller movies.

```python
sparql = """SELECT ?m_title
        WHERE {
            ?m <https://schema.org/hasGenre> <https://schema.org/Thriller> .
            ?m <https://schema.org/hasGenre> <https://schema.org/Crime> .
            ?m <https://schema.org/mytitle> ?m_title .

        }"""

res = g.query(sparql)
for row in res:
    print(f"Movie : {row.m_title}")
```

Actors older than 51

```python
sparql = """SELECT DISTINCT ?a_name ?age
        WHERE {
            ?a <https://schema.org/isActor> ?m .
            ?a <https://schema.org/myage> ?age .
            ?a <https://schema.org/myname> ?a_name .
            FILTER (?age > 51)
        }
        """

res = g.query(sparql)
for row in res:
    print(row.a_name, row.age)
```

Optional Graph Patterns

```python
sparql = """SELECT ?p_name ?p_bio ?p_photo
    WHERE {
        ?p a <https://schema.org/myPerson> .
        ?p <https://schema.org/myname> ?p_name .
        OPTIONAL {?p <https://schema.org/mybiography> ?p_bio}
        OPTIONAL {?p <https://schema.org/myphoto> ?p_photo}
    }"""

res = g.query(sparql)

for row in res:
    print(row.asdict())
```

Alternatives

```python
sparql = """SELECT ?m
        WHERE {
            {?m <https://schema.org/hasActor> <https://schema.org/al_pacino>}
            UNION
            {?m <https://schema.org/hasActor> <https://schema.org/jean_reno>}
        }"""

res = g.query(sparql)
for row in res:
    print(row.m)
```

Conjunction

```python
sparql = """SELECT ?m
        WHERE {
            ?m <https://schema.org/hasActor> <https://schema.org/al_pacino> .
            ?m <https://schema.org/hasActor> <https://schema.org/robert_de_niro>
        }"""

res = g.query(sparql)
for row in res:
    print(row.m)
```

Construct

```python
sparql = """CONSTRUCT {
        ?a <https://schema.org/isActor> ?m
        }
        WHERE { ?m <https://schema.org/hasActor> ?a }"""

res = g.query(sparql)
#g.update(sparql)

for row in res:
    print(row)
```

ASK

```python
sparql = """ASK {
    ?x <https://schema.org/myage> 67 ;
       <https://schema.org/myname> \"Tom Hanks\"
}"""
res = g.query(sparql)
for row in res:
    print(row)
```

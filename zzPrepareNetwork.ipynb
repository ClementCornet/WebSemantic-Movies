```python
import rdflib
g = rdflib.Graph()
g.parse('movies_project.owl')
iri = "http://semanticweb.org/cleme/projet#"
```

```python
#test_name = "Brad Pitt"
#test_id = 'brad_pitt'
test_id = 'al_pacino'
sparql = f"""SELECT ?n (COUNT(?n) as ?nCount)
    WHERE {{
        <{iri}{test_id}> <{iri}isActorOf> ?m .
        ?m <{iri}hasActor> ?a .
        BIND (datatype(?a) AS ?string) .
        FILTER (?a != <{iri}{test_id}>) .
        ?a <{iri}name> ?n
    }}
    GROUP BY ?n"""
res = g.query(sparql)
for row in res:
    #print(f"\n{row.m}")
    print(f"{row.n} | {row.nCount}")
```

```python
people = []
people_ids = {}
```

```python
sparql = f""" SELECT ?n 
    WHERE {{
        ?a a <{iri}Person> .
        ?a <{iri}name> ?n .
        BIND (datatype(?n) AS ?string) .
    }}
"""
res = g.query(sparql)
for i,row in enumerate(res):
    people.append((i+1, {'name':str(row.n)}))
    people_ids[str(row.n)] = i+1
people
```

```python
import networkx as nx
G = nx.Graph()
G.add_nodes_from(people)
#G.add_nodes_from([
#    (1, {'name':'Al Pacino'}),
#    (2, {'name':'Robert De Niro'}),
#    (3, {'name':'Val Kilmer'})
#])
#G.add_edge(1,2,weight=2)
#G.add_edge(1,3,weight=1)
#G.add_edge(2,3,weight=1)
```

Load the ontology and displays all the Persons (without using queries, without inference).

```python
from owlready2 import *

onto2 = get_ontology("movies_project.owl").load()
with onto2:
    people = onto2.search(is_a=onto2.myPerson)
    for p in people:
        print(p.iri,end='\n') # Display IRI
```

Load the ontology and displays all the Persons (using a query, without inference). Create the used query in text file under the data folder.

```python
import rdflib
g = rdflib.Graph()
g.parse("movies_project.owl")
sparql = """ SELECT ?p WHERE { ?p a  <https://schema.org/myPerson>}"""
res = g.query(sparql)
for row in res:
    print(row.p) # Display IRI
```

Loads the ontology and displays all the Actors (without using queries, using inference).

```python
onto3 = get_ontology("movies_project.owl").load()
with onto3:
    # Need to restart Jupyter Kernel to sync the reasonner a 2nd time
    sync_reasoner_pellet(infer_property_values=True, infer_data_property_values=True)
    actors = onto3.search(is_a=onto3.myPerson, rdfs_subclassof=onto3.Actor)
```

```python
for a in actors:
    print(a.iri)
```

Display all persons that are actors and directors

```python
sparql = """SELECT DISTINCT ?a 
        WHERE {
                ?a <https://schema.org/isActor> ?m1 .
                ?a <https://schema.org/isDirector> ?m2
                
                }"""

res = g.query(sparql)
sum(1 for _ in res)
for row in res:
    print(row.a)
```
